```python
CREATE TABLE Employee (
    Name VARCHAR(50),
    Department VARCHAR(50)
);

CREATE TABLE Manager (
    ManagerName VARCHAR(50),
    SubordinateName VARCHAR(50)
);

CREATE TABLE Colleague (
    EmployeeName1 VARCHAR(50),
    EmployeeName2 VARCHAR(50)
);

INSERT INTO Employee (Name, Department) VALUES ('john', 'engineering');
INSERT INTO Employee (Name, Department) VALUES ('susan', 'engineering');
INSERT INTO Employee (Name, Department) VALUES ('michael', 'marketing');
INSERT INTO Employee (Name, Department) VALUES ('emma', 'marketing');
INSERT INTO Employee (Name, Department) VALUES ('alex', 'finance');
INSERT INTO Employee (Name, Department) VALUES ('lisa', 'finance');
INSERT INTO Employee (Name, Department) VALUES ('jake', 'finance');
INSERT INTO Employee (Name, Department) VALUES ('james', 'sales');
INSERT INTO Employee (Name, Department) VALUES ('sarah', 'sales');

INSERT INTO Manager (ManagerName, SubordinateName) VALUES ('john', 'susan');
INSERT INTO Manager (ManagerName, SubordinateName) VALUES ('michael', 'emma');
INSERT INTO Manager (ManagerName, SubordinateName) VALUES ('alex', 'lisa');
INSERT INTO Manager (ManagerName, SubordinateName) VALUES ('alex', 'jake');
INSERT INTO Manager (ManagerName, SubordinateName) VALUES ('james', 'sarah');

-- Insert data into Colleague table
INSERT INTO Colleague (EmployeeName1, EmployeeName2)
SELECT e1.Name, e2.Name
FROM Employee e1
JOIN Employee e2 ON e1.Department = e2.Department AND e1.Name <> e2.Name;

-- 1. Who are the team members of John?
SELECT SubordinateName
FROM Manager
WHERE ManagerName = 'john';

-- 2. Who are the colleagues of Susan?
SELECT EmployeeName2
FROM Colleague
WHERE EmployeeName1 = 'susan';

-- 3. Who is the manager of Emma?
SELECT ManagerName
FROM Manager
WHERE SubordinateName = 'emma';

-- 4. Who are the team members of Alex?
SELECT SubordinateName
FROM Manager
WHERE ManagerName = 'alex';

-- 5. Who are the colleagues of James?
SELECT EmployeeName2
FROM Colleague
WHERE EmployeeName1 = 'james';

-- 6. Who are the department members of Finance?
SELECT Name
FROM Employee
WHERE Department = 'finance';

-- 7. Is Susan a colleague of James?
SELECT COUNT(*)
FROM Colleague
WHERE (EmployeeName1 = 'susan' AND EmployeeName2 = 'james')
   OR (EmployeeName1 = 'james' AND EmployeeName2 = 'susan');

-- 8. Is Michael a manager of Sarah?
SELECT COUNT(*)
FROM Manager
WHERE ManagerName = 'michael' AND SubordinateName = 'sarah';
```
